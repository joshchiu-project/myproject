```python
my_data = pd.read_csv(my_file,index_col=0)#中文编码
#store neg
my_data_test_lable = my_data['is_neg']

#preprocess
my_data.drop(['is_resid_foreign_country','first_pay_diff_estab','first_gather_diff_estab','living_day','total_expotrt_price_cnt','total_import_price_cnt','foreign_company_trans_cnt','is_neg','contractamount_dollar_sum'],axis=1,inplace=True)
my_data_test_1 = my_data

```

# k-mean


```python
from sklearn.metrics import silhouette_score
Scores = []  # 存放轮廓系数
k = 2

estimator = KMeans(n_clusters=k)  # 构造聚类器
estimator.fit(my_array)
Scores.append(silhouette_score(my_array,estimator.labels_,metric='euclidean'))
```
